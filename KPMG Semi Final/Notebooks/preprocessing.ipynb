Load the dataset

```python
import pandas as pd 
import numpy as np 
import openpyxl

path = '../Datasets/DamDataExcel.xlsx'
df = pd.read_excel(path)
```

```python
df.head()
```

```python
df.set_index('_id', inplace=True);df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date']);df.info()
```

```python
for i in df.columns[1:].tolist():
    df[i] = pd.to_numeric(df[i].str.replace(',', '.').str.replace('%', ''))
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].dt.strftime("%d-%m-%Y");df.head()
```

```python
df.to_csv("../Datasets/dataframe.csv", index=False)
```
