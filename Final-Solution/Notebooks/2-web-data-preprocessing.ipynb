```python
otopark = pd.read_csv('../Datasets/EXTERNAL/ispark_parking.csv')
otopark.head()
```

```python
otopark['COUNTY_NAME'].unique()
```

```python
otopark = otopark[['PARK_TYPE_DESC', 'CAPACITY_OF_PARK', 'WORKING_TIME','COUNTY_NAME','LONGITUDE','LATITUDE']]
otopark.columns = ['park_tipi', 'kapasite', 'calisma_saati', 'ilce', 'longitude', 'latitude']
otopark.head(3)
```

```python
otopark[otopark['calisma_saati']=='24 Saat'] = '00:00-23:59'
otopark['calisma_saati'].unique()
```

```python
otopark[['acilis_saati', 'kapanis_saati']] = otopark['calisma_saati'].str.split('-', expand=True)
otopark.loc[otopark['kapanis_saati']=='00:00', 'kapanis_saati'] = '23:59'
otopark.head(3)
```

```python
import datetime as dt

otopark['acilis_saati'] = otopark['acilis_saati'].apply(lambda x: dt.datetime.strptime(x.strip(), '%H:%M').time())
otopark['kapanis_saati'] = otopark['kapanis_saati'].apply(lambda x: dt.datetime.strptime(x.strip(), '%H:%M').time())
otopark.head(3)
```

```python
calisma_suresi = []
for row in range(len(otopark)):
    acilis = otopark['acilis_saati'][row].hour
    kapanis = otopark['kapanis_saati'][row].hour
        
    if acilis > kapanis:
        calisma_suresi.append(acilis - kapanis + 12)
    
    else:
        calisma_suresi.append(kapanis - acilis)
otopark['calisma_suresi'] = calisma_suresi
otopark.head(3)
```

```python
calisma_tipi = []

for row in range(len(otopark)):
    if otopark.iloc[row]['calisma_suresi'] >= 17:
        calisma_tipi.append('tum_gun')
    elif dt.time(6,0,0) <= otopark.iloc[row]['acilis_saati'] <= dt.time(19,0,0):
        calisma_tipi.append('gunduz')
    elif dt.time(18,0,0) < otopark.iloc[row]['acilis_saati'] or otopark.iloc[row]['acilis_saati'] < dt.time(6,0,0):
        calisma_tipi.append('gece')
    else:
        calisma_tipi.append('bilinmiyor')

    
otopark['calisma_tipi'] = calisma_tipi
```

```python
otopark.head(3)
```

```python
otopark['calisma_tipi'].value_counts()
```

```python
otopark = otopark.drop(['calisma_saati','ilce'], axis=1)
otopark.to_csv('../Datasets/PROCESSED/otopark.csv', index=False)
```

```python
with open('../Datasets/EXTERNAL/raylı-sistem-yolcu-segmenti.csv', encoding='iso-8859-9') as f:
    lines = f.readlines()
    
data = [line.strip().split(',') for line in lines]
```

```python
columns = ["passage_cnt","passanger_cnt","transaction_year",
           "transaction_month","transaction_day","line",
           "station_name","station_number","town",
           "age","longitude","latitude"]

metro = pd.DataFrame(data[1:], columns=columns)
metro.head(2)
```

```python
metro.drop(['line'], axis=1, inplace=True)
```

```python
metro.head(12)
```

```python
metro['transaction_year'] = metro['transaction_year'].astype(int)
metro['transaction_month'] = metro['transaction_month'].astype(int)
metro['transaction_day'] = metro['transaction_day'].astype(int)


metro['transaction_datetime'] = metro.apply(
    lambda x: dt.datetime(x['transaction_year'], x['transaction_month'], x['transaction_day']), axis=1)
```

```python
metro.head(3)
```

```python
metro = metro[['passage_cnt','passanger_cnt','age','longitude','latitude','transaction_datetime']]
metro.columns = ['gecis_sayisi','yolcu_sayisi','yas_grubu','longitude','latitude','tarih']
metro.head(3)
```

```python
metro['longitude'] = metro['longitude'].str.replace('"', '')
metro['latitude'] = metro['latitude'].str.replace('"', '')
metro.head(3)
```

```python
metro.isnull().sum()
```

```python
empty_longitude = metro[metro['longitude'] == '']
empty_latitude = metro[metro['latitude'] == '']

metro.loc[metro['longitude'] == '', 'longitude'] = 0.0
metro.loc[metro['latitude'] == '', 'latitude'] = 0.0
```

```python
metro['gecis_sayisi'] = metro['gecis_sayisi'].astype(int)
metro['yolcu_sayisi'] = metro['yolcu_sayisi'].astype(int)

metro['longitude'] = pd.to_numeric(metro['longitude'], errors='coerce')
metro['latitude'] = pd.to_numeric(metro['latitude'], errors='coerce')

metro.head(3)
```

```python
metro.to_csv('../Datasets/PROCESSED/metro.csv', index=False)
```
